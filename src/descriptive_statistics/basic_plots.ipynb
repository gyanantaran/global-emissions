```python
# %% Data Visualisation

# Set the style for Seaborn
sns.set(style="whitegrid")
```

```python
# Visualize relationships between variables
sns.pairplot(emissions_df)
plt.show()
```

```python
# %% Visualization 1: Time Trend of Global Emissions

plt.figure(figsize=(12, 6))
sns.lineplot(x="Year", y="Emissions.Production.CO2.Total", data=emissions_df)
plt.title("Time Trend of Global CO2 Emissions")
plt.xlabel("Year")
plt.ylabel("Total CO2 Emissions (kilotons)")
plt.show()
```

```python
# %% Visualization 2: Distribution of CO2 Emissions

plt.figure(figsize=(12, 6))
sns.histplot(emissions_df["Emissions.Production.CO2.Total"], bins=30, kde=True)
plt.title("Distribution of Total CO2 Emissions")
plt.xlabel("Total CO2 Emissions (kilotons)")
plt.ylabel("Frequency")
plt.show()
```

```python
# %% Visualization 3: Box Plot of GDP across Countries
plt.figure(figsize=(12, 6))
sns.boxplot(x="Country.Name", y="Country.GDP", data=emissions_df)
plt.title("Box Plot of GDP across Countries")
plt.xlabel("Country")
plt.ylabel("GDP")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
plt.show()
```

```python
# %% Visualization 4: Correlation Heatmap
# Select numerical columns for correlation analysis
plt.figure(figsize=(10, 8))
numerical_columns = emissions_df.select_dtypes(include=["float64", "int64"]).columns
correlation_matrix = emissions_df[numerical_columns].corr()

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.savefig(docs_dir + "/correlation-matrix.png")
plt.show()
```

```python
# %% Visualization 5: Global Share of CO2 Emissions by Source
source_columns = [
    "Emissions.Global Share.CO2.Cement",
    "Emissions.Global Share.CO2.Coal",
    "Emissions.Global Share.CO2.Gas",
    "Emissions.Global Share.CO2.Oil",
    "Emissions.Global Share.CO2.Flaring",
]
```

```python
df_global_share = emissions_df[source_columns]

plt.figure(figsize=(10, 6))
df_global_share.mean().plot(kind="bar", color="skyblue")
plt.title("Global Share of CO2 Emissions by Source")
plt.xlabel("Emission Source")
plt.ylabel("Global Share (%)")
plt.xticks(rotation=45, ha="right")
plt.show()
```
