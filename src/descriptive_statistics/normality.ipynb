```python
from src.utils.data_loader import emissions_df

import numpy as np 
from matplotlib import pyplot as plt
from scipy.stats import probplot
```

```python
year = np.random.randint(high=2018, low=1992)
year_col = emissions_df['Year']

data = emissions_df[year_col==year]

var_col = 'Emissions.Global Share.CO2.Cement'
var_data = (data[var_col]).to_list()

probplot(var_data, plot=plt)
plt.title(f'Q-Q: Source {var_col.split(".") [-1]} {year}')
plt.show()
```
