```python
from src.utils.data_loader import emissions_df
# from src.descriptive_statistics.summary import top_countries_mean_global_share_emissions
from src.descriptive_statistics.summary import top_countries_mean_production_emissions
from src.utils.paths import docs_dir

from matplotlib import pyplot as plt
```

```python
# top_10_countries_mean_global_share = top_countries_mean_global_share_emissions.keys().tolist()
top_10_countries_mean_production = top_countries_mean_production_emissions.keys().tolist()
```

```python
# Create a line plot for each country
plt.figure(figsize=(12, 8))
# for country in top_10_countries_mean_global_share:
for country in top_10_countries_mean_production:
    country_data = emissions_df[emissions_df['Country.Name'] == country]
    plt.plot(country_data['Year'], country_data['Emissions.Production.CO2.Total'], label=country)

# Customize the plot
plt.title('Global Share of CO2 Emissions Over the Years - Top 10 Countries')
plt.xlabel('Year')
plt.ylabel('The total amount of CO2 produced (kilotons)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.savefig(docs_dir + '/summary_plots/producers.png')
plt.show()
```
